## Mental Health - Depression Screener 2015-2016 (DPQ_I)

```python
#load data:
mental_health_df_15=pd.read_sas('./2015datasets/DPQ_I.XPT')
mental_health_df_15.head()
```

```python
# number of rows:
mental_health_df_15.shape[0]
```

```python
# drop last row
mental_health_df_15 = mental_health_df_15.drop(columns=['DPQ100'])
mental_health_df_15.head()
```

```python
# drop rows with missing values
mental_health_df_15=mental_health_df_15.dropna()
```

```python
# number of rows
mental_health_df_15.shape[0]
```

```python
# replace 7 and 9 with none and drop those rows
mental_health_df_15 = mental_health_df_15.replace({7: None})
mental_health_df_15 = mental_health_df_15.replace({9: None})
mental_health_df_15=mental_health_df_15.dropna()
mental_health_df_15.shape[0]
```

```python
# convert data type into int64
mental_health_df_15['DPQ010'] = mental_health_df_15['DPQ010'].astype(np.int64)
mental_health_df_15['DPQ020'] = mental_health_df_15['DPQ020'].astype(np.int64)
mental_health_df_15['DPQ030'] = mental_health_df_15['DPQ030'].astype(np.int64)
mental_health_df_15['DPQ040'] = mental_health_df_15['DPQ040'].astype(np.int64)
mental_health_df_15['DPQ050'] = mental_health_df_15['DPQ050'].astype(np.int64)
mental_health_df_15['DPQ060'] = mental_health_df_15['DPQ060'].astype(np.int64)
mental_health_df_15['DPQ070'] = mental_health_df_15['DPQ070'].astype(np.int64)
mental_health_df_15['DPQ080'] = mental_health_df_15['DPQ080'].astype(np.int64)
mental_health_df_15['DPQ090'] = mental_health_df_15['DPQ090'].astype(np.int64)
```

```python
# sum the scores of each questions
cols = ['DPQ010','DPQ020','DPQ030','DPQ040','DPQ050','DPQ060','DPQ070','DPQ080','DPQ090']
mental_health_df_15['SCORE'] = mental_health_df_15[cols].sum(axis=1)
mental_health_df_15.head()
```

```python
mental_health_df_15.columns
```

```python
# drop other rows
mental_health_df_15=mental_health_df_15[['SEQN','DPQ010', 'DPQ020', 'DPQ030', 'DPQ040', 'DPQ050', 'DPQ060',
       'DPQ070', 'DPQ080', 'DPQ090','SCORE']]
mental_health_df_15.head()
```

```python
mental_health_df_15['SCORE'].describe()
```

```python
# categorize score as “none or minimum” (0–4), “mild” (5–9), “moderate” (10–14), “moderately severe” (15–19), and “severe” (20–27)
# 0 for none, 1 for mild, 2 for moderate, 3 for moderately severe, 4 for severe
#https://doi.org/10.1016/j.jad.2020.03.031
mental_health_df_15['DEPRESSION']=pd.cut(mental_health_df_15['SCORE'], bins=[0, 4, 9, 14, 19, 27], include_lowest=True, labels=['0', '1', '2', '3','4'])
mental_health_df_15
```

```python
#CHECKING NAN VALUES
mental_health_df_15.isnull().sum()
```

```python
# load data
fasting_glucose_df_15=pd.read_sas('./2015datasets/GLU_I.XPT')
fasting_glucose_df_15.head()
```

```python
# number of rows
fasting_glucose_df_15.shape[0]
```

```python
# drop rows with missing data
fasting_glucose_df_15=fasting_glucose_df_15.dropna()
# number of rows
fasting_glucose_df_15.shape[0]
```

```python
# drop LBDGLUSI and WTSAFPRP column
fasting_glucose_df_15=fasting_glucose_df_15[['SEQN','LBXGLU']]
fasting_glucose_df_15.head()
```

```python
# diabetes function
# 1 if diabetic
# 0 if non diabetic 
def check_for_diabetes_for_fasting_glucose(fasting_glucose):
  if pd.notna(fasting_glucose):
    if fasting_glucose>=126:
      return 1
    else:
      return 0;
  return np.nan
```

```python
# apply function

fasting_glucose_df_15['Diabetes']=fasting_glucose_df_15.apply(lambda x: check_for_diabetes_for_fasting_glucose(fasting_glucose=x['LBXGLU']), 1)
fasting_glucose_df_15.head()
```

```python
# drop LBXGLU
fasting_glucose_df_15=fasting_glucose_df_15[['SEQN','Diabetes']]
fasting_glucose_df_15.head()
```

```python
# diabetes function
# 1 if diabetic
# 0 if non diabetic 
def check_for_diabetes_for_aic_value(aic_value):
  if pd.notna(aic_value):
    if aic_value>=6.5:
      return 1
    else:
      return 0;
  return np.nan
```

```python
# apply function
df=pd.DataFrame()
aic_test_df_15['Diabetes']=aic_test_df_15.apply(lambda x: check_for_diabetes_for_aic_value(aic_value=x['LBXGH']), 1)
aic_test_df_15.head()
```

```python
# drop LBXGH
aic_test_df_15=aic_test_df_15[['SEQN','Diabetes']]
aic_test_df_15.head()
```

```python
# concat two dataframes
frames = [aic_test_df_15, fasting_glucose_df_15]
diabetes_df_15 = pd.concat(frames)
```

```python
diabetes_df_15.head()
```

```python
diabetes_df_15 = diabetes_df_15.drop_duplicates('SEQN')
diabetes_df_15['Diabetes'].value_counts()[1]
```

```python
#Rename Column Name:
diabetes_df_15.columns = diabetes_df_15.columns.str.replace('Diabetes', 'DIABETES')
diabetes_df_15
```

```python
# CHECKING NAN VALUES
diabetes_df_15.isnull().sum()
```

```python
# load data
body_measures_df_15=pd.read_sas('./2015datasets/BMX_I.XPT')
body_measures_df_15.head()
```

```python
# number of rows
body_measures_df_15.shape[0]
```

```python
# dropping other rows
body_measures_df_15=body_measures_df_15[['SEQN','BMXWT','BMXHT','BMXBMI']]
body_measures_df_15.head()
```

```python
# bmi function
def calculate_bmi(height,weight,existing_value):
  if existing_value is None:
    height=height/100
    return (weight/(height*height))
  else:
    return existing_value
```

```python
df = pd.DataFrame()
body_measures_df_15['BMXBMI']=body_measures_df_15.apply(lambda x: calculate_bmi(height=x['BMXHT'], weight=x['BMXWT'], existing_value=x['BMXBMI']), 1)
body_measures_df_15.head()
```

```python
# drop rows with missing data
body_measures_df_15=body_measures_df_15.dropna()
# number of rows
body_measures_df_15.shape[0]
```

```python
body_measures_df_15.head()
```

```python
# RENAMING COLUMN NAME:
body_measures_df_15.columns = body_measures_df_15.columns.str.replace('BMXWT', 'WEIGHT')
body_measures_df_15.columns = body_measures_df_15.columns.str.replace('BMXHT', 'HEIGHT')  
body_measures_df_15.columns = body_measures_df_15.columns.str.replace('BMXBMI', 'BMI') 
body_measures_df_15
```

```python
# load audiometry data 
audiometry_df_15=pd.read_sas('./2015datasets/AUX_I.XPT')
# number of audiometry
audiometry_df_15.shape[0]
```

```python
audiometry_df_15.head()
```

```python
# dropped columns that are not required
audiometry_df_15=audiometry_df_15[['SEQN','AUXU500R','AUXU1K1R','AUXU1K2R','AUXU2KR','AUXU3KR','AUXU4KR','AUXU6KR','AUXU500L','AUXU1K1L',	'AUXU1K2L','AUXU2KL','AUXU3KL','AUXU4KL','AUXU6KL']]
audiometry_df_15.head()
```

```python
# create columns with averages of various frequencies
audiometry_df_15['Right500'] = audiometry_df_15[['AUXU500R']]
audiometry_df_15['Right1000'] = audiometry_df_15[['AUXU1K1R','AUXU1K2R']].mean(axis=1)
audiometry_df_15['Right2000'] = audiometry_df_15[['AUXU2KR']]
audiometry_df_15['Right3000'] = audiometry_df_15[['AUXU3KR']]
audiometry_df_15['Right4000'] = audiometry_df_15[['AUXU4KR']]
audiometry_df_15['Right6000'] = audiometry_df_15[['AUXU6KR']]
audiometry_df_15['Left500'] = audiometry_df_15[['AUXU500L']]
audiometry_df_15['Left1000'] = audiometry_df_15[['AUXU1K1L','AUXU1K2L']].mean(axis=1)
audiometry_df_15['Left2000'] = audiometry_df_15[['AUXU2KL']]
audiometry_df_15['Left3000'] = audiometry_df_15[['AUXU3KL']]
audiometry_df_15['Left4000'] = audiometry_df_15[['AUXU4KL']]
audiometry_df_15['Left6000'] = audiometry_df_15[['AUXU6KL']]
audiometry_df_15.head()
```

```python
# drop columns other than columns values
audiometry_df_15=audiometry_df_15[['SEQN','Right500','Right1000','Right2000','Right3000','Right4000','Right6000','Left500','Left1000','Left2000','Left3000','Left4000','Left6000']]
audiometry_df_15.head()
```

```python
# calculate SFHL and HFHL for each eear
audiometry_df_15['LeftSFHL'] = audiometry_df_15[['Left500','Left1000','Left2000','Left4000']].mean(axis=1)
audiometry_df_15['LeftHFHL'] = audiometry_df_15[['Left3000','Left4000','Left6000']].mean(axis=1)
audiometry_df_15['RightSFHL'] = audiometry_df_15[['Right500','Right1000','Right2000','Right4000']].mean(axis=1)
audiometry_df_15['RightHFHL'] = audiometry_df_15[['Right3000','Right4000','Right6000']].mean(axis=1)
audiometry_df_15.head()
```

```python
# remove other columns
audiometry_df_15=audiometry_df_15[['SEQN','LeftSFHL','LeftHFHL','RightSFHL','RightHFHL']]
audiometry_df_15.head()
```

```python
# number of rows with missing data
audiometry_df_15.isnull().any(axis=1).sum()
```

```python
# total number of rows
audiometry_df_15.shape[0]
```

```python
# drop rows with missing data
audiometry_df_15=audiometry_df_15.dropna()
# total number of rows
audiometry_df_15.shape[0]
```

```python
audiometry_df_15.head()
```

```python
# find SFHL and HFHL of the better ear
audiometry_df_15["SFHL"]=audiometry_df_15[["LeftSFHL", "RightSFHL"]].min(axis=1)
audiometry_df_15["HFHL"]=audiometry_df_15[["LeftHFHL", "RightHFHL"]].min(axis=1)
audiometry_df_15.head()
```

```python
# drop other rows
audiometry_df_15=audiometry_df_15[['SEQN','SFHL','HFHL']]
audiometry_df_15.head()
```
