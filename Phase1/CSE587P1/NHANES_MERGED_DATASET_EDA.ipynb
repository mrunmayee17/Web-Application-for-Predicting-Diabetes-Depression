## GRAPHICAL EDA

```python
# Dropping some columns of DS to get heat map, and to understand correlation better.
DS_COR = DS[['GENDER', 'AGE', 'HISPANIC_O', 'SLEEP_HRS', 'DIASTOLIC',
       'SYSTOLIC', 'WEIGHT', 'HEIGHT', 'BMI', 'DIABETES', 'SCORE',
       'DEPRESSION', 'SFHL', 'HFHL']]
DS_COR
```

```python
# 7. correlation map of the data set.
#To understand the relationship between any two features in our data set, we used correlation matrix. According to this correlation matrix, we were able to derive some inferences such as the relationship between age and BMI(Body to Mass Index).
fig, ax = plt.subplots(figsize=(15,15))  
sns.heatmap(DS_COR.corr(), cmap="YlGnBu", annot=True, ax=ax, linewidths=.5)
plt.show()
```

```python
# 8. Plotting Distributions for better understandings:
# Histogram features provide strong indications of the proper distributional model for the data.
# E.g. We can infer from the histogram that the average height of a person is between 160 to 170cm. Average sleep hours are between 8 to 9 hours.

plt.figure(figsize=(26, 26))
for i, col in enumerate(DS.columns[1:]):
    plt.subplot(3, 4, i+1)
    plt.hist(DS[col], color='red', bins=100)
    plt.hist(DS[col], color='green', bins=100)
    plt.title(col)

```

```python
DS.columns
```

```python
# 9. Plotting Box Plot for better understandings :
# To understand outlier in the dataset, we use box plot. By default, the boxplot chart excludes any values outside of the 1.5 IQR (Interquartile Range) range, which eliminates the outlier values.
# Box plot crucially identified the outliers in speech frequency hearing loss and high frequency hearing loss. This will help in dealing with the outliers in phase 2.

DS.boxplot(column=['GENDER', 'AGE', 'HISPANIC_O', 'SLEEP_HRS', 'DIASTOLIC',
       'SYSTOLIC', 'WEIGHT', 'HEIGHT', 'BMI', 'DIABETES', 'SCORE',
       'DEPRESSION', 'SFHL', 'HFHL'], figsize=(20,30))  
```

```python
# Trying to visualize indivisual box plot:

ax = sns.boxplot(x = DS[DS['HISPANIC_O']== 3.0]['DIABETES'])
ax.axes.set_title("", fontsize=20, y=1.01)
plt.xlabel("DS['DIABETES']", labelpad=14);
```

```python
# 10. Trying to visualize scatter plots between  sleep and diabetes
# We tried to visualize the correlations which we found in the correlation matrix â€“ heat map here. It is helpful in determining the relationship between sleep and diabetes and diabetes and BMI which will be helpful in the modeling phase.
plt.figure(figsize=(10,10))
plt.scatter(x='SLEEP_HRS',y='HISPANIC_O',data=DS)
plt.xlabel('SLEEP')
plt.ylabel('DIABETES')
```

```python
# Trying to visualize scatter plots between diabetes and Etnicity:
# We tried to visualize the correlations which we found in the correlation matrix â€“ heat map here. It is helpful in determining the relationship between sleep and diabetes and diabetes and BMI which will be helpful in the modeling phase.

plt.figure(figsize=(10,10))
plt.scatter(x='DIABETES',y='HISPANIC_O',data=DS)
plt.xlabel('SLEEP_HRS')
plt.ylabel('GENDER')
```

```python
# Trying to visualize scatter plots between diabetes and Etnicity:
plt.figure(figsize=(10,10))
plt.scatter(x='DIABETES',y='BMI',data=DS)
plt.xlabel('DIABETES')
plt.ylabel('BMI')
```
